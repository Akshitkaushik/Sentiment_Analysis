```python
from tkinter import *
from tkinter import filedialog
from tkinter import simpledialog as sd
from tkinter import messagebox as msg
import os
import pandas as pd
import numpy as np
import nltk
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB



#objects of classes
ps=PorterStemmer()
cv=CountVectorizer()
tf=TfidfVectorizer()
lr=LogisticRegression(solver='lbfgs')
gb=GaussianNB()
mn=MultinomialNB()

root=Tk()
bgclr='teal'
root.state('zoomed')
root.update()
root.configure(bg=bgclr)
root.resizable(width=False,height=FALSE)
lbl_head=Label(root,bg=bgclr,text='Natural Processing Language Analysis',font=('Times',30,'bold'))
lbl_head.pack(side='top',anchor='c')

#login function to called from home button
def login(frm,e1,e2):
    ent_user=e1.get()
    ent_pass=e2.get()
    
    if(len(ent_user)==0 or len(ent_pass)==0):
        msg.showwarning('Validation Problem','Please Fill All Fields')
    else:
        if(ent_user=='admin'and ent_pass=='admin'):
            msg.showinfo('Login','Welcome Admin')
            frm.destroy()
            WelcomeAdmin()
        else:
            msg.showerror('Login Failed','Invalid username and password')

            

            
            

            
def logout(frm):
    frm.destroy()
    home()    
            
def back(frm):
    frm.destroy()
    WelcomeAdmin()
    

def browser():
    file_path=filedialog.askopenfile()
    entry_dir.insert(0,file_path.name)
    
    
    
# Login comment reset
def reset_login(e1,e2):
    u=e1.get()
    p=e2.get()
    e1.delete(0,len(u))
    e2.delete(0,len(p))

# Code to reset input comments inside frames  
def reset():
    #reset dir path
    s=entry_dir.get()
    ln=len(s)
    entry_dir.delete(0,ln)
    #reset keyword
    p=entry_key.get()
    ln2=len(p)
    entry_key.delete(0,ln2)    
    
# Funtion to clean text
    
def clean_txt(text):
    sp=string.punctuation
    sw=stopwords.words('English')
    sw.remove('not')
    new_text=re.sub(f'[{sp}]',"",text)
    new_text=new_text.lower()
    words=word_tokenize(new_text)
    new_words=[]
    for w in words:
        if (w not in sw):
            nw=ps.stem(w)
            new_words.append(nw)
    return " ".join(new_words)    
    
#Restaurant Review NLP code 
    
def check_sentiment(frm):
    file=entry_dir.get()
    new_comment=entry_key.get()
   

    df=pd.read_csv(file,delimiter='\t',quoting=3)
    df['Review']=df.Review.map(clean_txt)
    cleaned_txt=clean_txt(new_comment)
    
    cv=CountVectorizer()
    X=cv.fit_transform(df.Review).toarray()
    transformed_comment=cv.transform([cleaned_txt]).toarray()
    y=df.Liked
    
    lr.fit(X,y,)
    
    result_comment=lr.predict(transformed_comment)
        
    if result_comment==1:
        show='Like'
        msg.showinfo('Review',show)
    else:
        show='Not Like'
        msg.showinfo('Review',show)
    
    
    
# Resturant review frame code
            
def rest_review(frm):
    frm.destroy()
    
    login_frm=Frame(root,bg=bgclr)
    login_frm.place(x=0,y=100,width=root.winfo_width(),height=root.winfo_height())
   
    
    global entry_dir
    global entry_key
    
   

    entry_dir=Entry(login_frm,font=('',14),bd=3)
    entry_dir.place(x=250,y=150)
    
    entry_key=Entry(login_frm,font=('',14),bd=3)
    entry_key.place(x=250,y=200)  
    
    btn_rst=Button(login_frm,command=reset,text=" Reset ",font=('book antiqua',13),bd=3)
    btn_rst.place(x=500,y=250)   
       
    back_btn=Button(login_frm,width=10,command=lambda:(back(login_frm)),font=('',12,'bold'),text='Back',bd=5)
    back_btn.place(relx=.03,y=10)
    
    logout_btn=Button(login_frm,width=15,command=lambda:(logout(login_frm)),font=('',12,'bold'),text='Logout',bd=5)
    logout_btn.place(relx=.85,y=100)
    
    lbl_title=Label(login_frm,text='Twitter Detection',font=('cambria',24,'bold','underline'),bg='teal',fg='white')
    lbl_title.pack(padx=10)

    lbl_dir=Label(login_frm,text="Upload Source Data: ",font=('cambria',14,'bold'),bg='teal',fg='white')
    lbl_dir.place(x=50,y=150)

    lbl_key=Label(login_frm,text="New Comment: ",font=('cambria',14,'bold'),bg='teal',fg='white')
    lbl_key.place(x=50,y=200)
    
    btn_browser=Button(login_frm,text='Browser',command=browser,font=('',13),bd=5)
    btn_browser.place(x=500,y=150)
    
    btn_srch=Button(login_frm,text='Check sentiment',command=lambda:check_sentiment(login_frm),font=('',13),bd=3)
    btn_srch.place(x=250,y=250)       

    
    
#Twitter Review NLP code 
    
def twitter_clean(frm):
    file=entry_dir.get()
    new_comment=entry_key.get()
   

    df=pd.read_csv(file,delimiter='\t',quoting=3)
    df['SentimentText']=df.SentimentText.map(clean_txt)
    cleaned_txt=clean_txt(new_comment)
    

    X=tf.fit_transform(df.SentimentText).toarray()
    transformed_comment=tf.transform([cleaned_txt]).toarray()
    y=df.Sentiment
    
    gb.fit(X,y,)
    
    result_comment=gb.predict(transformed_comment)
    
    if result_comment==1:
        show='Happy'
        msg.showinfo('Result',show)

    else:
        show='Sad'
        msg.showinfo('Result',show)

    



# Twitter review frame code

    
def Twitter(frm):
    frm.destroy()
    
    
    login_frm=Frame(root,bg=bgclr)
    login_frm.place(x=0,y=100,width=root.winfo_width(),height=root.winfo_height())
    
    global entry_dir
    global entry_key
    
    entry_dir=Entry(login_frm,font=('',14),bd=3)
    entry_dir.place(x=250,y=150)
    
    entry_key=Entry(login_frm,font=('',14),bd=3)
    entry_key.place(x=250,y=200)  
    
    btn_rst=Button(login_frm,command=reset,text=" Reset ",font=('book antiqua',13),bd=3)
    btn_rst.place(x=500,y=250)   
       
    back_btn=Button(login_frm,width=10,command=lambda:(back(login_frm)),font=('',12,'bold'),text='Back',bd=5)
    back_btn.place(relx=.03,y=10)
    
    logout_btn=Button(login_frm,width=15,command=lambda:(logout(login_frm)),font=('',12,'bold'),text='Logout',bd=5)
    logout_btn.place(relx=.85,y=100)
    
    lbl_title=Label(login_frm,text='Twitter Detection',font=('cambria',24,'bold','underline'),bg='teal',fg='white')
    lbl_title.pack(padx=10)

    lbl_dir=Label(login_frm,text="Upload Source Data: ",font=('cambria',14,'bold'),bg='teal',fg='white')
    lbl_dir.place(x=50,y=150)

    lbl_key=Label(login_frm,text="New Comment: ",font=('cambria',14,'bold'),bg='teal',fg='white')
    lbl_key.place(x=50,y=200)
    
    btn_browser=Button(login_frm,text='Browser',command=browser,font=('',13),bd=5)
    btn_browser.place(x=500,y=150)
    
    btn_srch=Button(login_frm,text='Check sentiment',command=lambda:twitter_clean(login_frm),font=('',13),bd=3)
    btn_srch.place(x=250,y=250)       
            
        


    
        
        
        
        
        
        
        
        
        
        
        
        
        


            
            
            
            
            
            
            
            
def WelcomeAdmin():
    
    login_frm=Frame(root,bg=bgclr)
    login_frm.place(x=0,y=100,width=root.winfo_width(),height=root.winfo_height())
    
    lbl_head=Label(login_frm,bg='teal',text='Sentiment Analysis',font=('cambria',24,'bold','underline'))
    lbl_head.pack(side='top',anchor='c')
    
    review_btn=Button(login_frm,text='Restaurant Review',command=lambda:(rest_review(login_frm)),bd=5,font=('',12,'bold'),width=40)
    review_btn.place(x=550,y=215)
   
    Twitter_btn=Button(login_frm,text='Twitter Response',command=lambda:(Twitter(login_frm)),bd=5,font=('',12,'bold'),width=40)
    Twitter_btn.place(x=550,y=300)
   
    logout_btn=Button(login_frm,text='logout',command=lambda:logout(login_frm),bd=5,font=('',12,'bold'),width=15)
    logout_btn.place(relx=.85,y=100)


    
                                                                             











def home():
    login_frm=Label(root,bg=bgclr)
    login_frm.place(x=0,y=100,width=root.winfo_width(),height=root.winfo_height())
    
    lbl_user=Label(login_frm,bg=bgclr,font=("",20,""),fg='white',text='Username')
    lbl_user.place(x=300,y=100)
    
    lbl_pass=Label(login_frm,bg=bgclr,font=(" ",20," "),fg="white",text='Password')
    lbl_pass.place(x=300,y=150)
    
    ent_user=Entry(login_frm,bd=5,font=('',15,''))
    ent_user.focus()
    ent_user.place(x=490,y=105)
    
    ent_pass=Entry(login_frm,show='*',bd=5,font=('',15,''))
    ent_pass.focus()
    ent_pass.place(x=490,y=155)
    
    lgn_btn=Button(login_frm,command=lambda:login(login_frm,ent_user,ent_pass),text='Login',bd=5,font=("",12,""))
    lgn_btn.place(x=500,y=215)
   
    rst_btn=Button(login_frm,text='Reset',command=lambda:reset_login(ent_user,ent_pass),bd=5,font=("",12,""))
    rst_btn.place(x=650,y=215)

    
    
    


    
    
    
    
    
    
    
    
home()
mainloop()










```
